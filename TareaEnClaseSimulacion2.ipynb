
# Ejercico 1
1 Con los datos de Fifa, organizar a los jugadores de acuerdo al peso en las siguientes escalas y generar un cuadro tipo PIE

Debajo 125 Lbs.
125-150
150-175
175 o superior

```python
#Trabajar con otro tipo de datos Fifa
fifa = pd.read_csv(io.StringIO(archivo_fifa['fifa_datos.csv'].decode('utf-8')))
```

```python
fifa["Weight"] = fifa["Weight"].replace({'lbs':''}, regex=True)
#fifa['Weight']=fifa.dropna(inplace=True)
fifa['Weight'] = fifa['Weight'].astype(float)
fifa.head(5)
```

Jugadores debajo 125 lb





```python
condicion1 = fifa.loc[fifa['Weight'] < 125 ]
condicion1.head(5)
```

Jugadores 125 -150 lbs

```python
condicion2 = fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] <= 150)]
condicion2.head(5)
```

Jugadores 150 - 175 lbs

```python
condicion3 = fifa.loc[(fifa['Weight'] > 150) & (fifa['Weight'] <= 175)]
condicion3.head(5)
```

Jugadores Superior a 175

```python
condicion4 = fifa.loc[fifa['Weight'] > 175 ]
condicion4.head(5)
```

```python
# Generar un grafico 
condicion1 = condicion1.count()[0]
condicion2 = condicion2.count()[0]
condicion3 = condicion3.count()[0]
condicion4 = condicion4.count()[0]

plt.figure()

etiquetas = ['< 125.', '125-150' ,'151-175','> 175']
colores = ['#9AFF33', '#FF4A33', '#D433FF', '#FF3371' ]
plt.pie([condicion1, condicion2,condicion3,condicion4], labels=etiquetas, colors=colores, autopct='%.2f %%')
plt.title(' jugadores de acuerdo al peso')
plt.show()
```
