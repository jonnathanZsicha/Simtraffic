```python
#LOADING DATA Poblacion
dfp= pd.read_csv('/content/1.Poblaciones.csv', error_bad_lines=False ,sep=';' , encoding='latin-1' ) 
dfp.head(5)
```

Descripcion de Variable dataset Poblacion

-Periodo : tipo Date Fecha del periodo realizado la encuesta

-Indicadores :Tipo de trabajo que realiza cierta poblacion

-Nacional : tipo Float total de la poblacion a nivel nacional

```python
# Import plotly and create a figure
import plotly.graph_objects as go
fig = go.Figure()



# Add the scatter traces to the figure
for column_header in ['Periodo']:
    fig.add_trace(go.Scatter(
        x=dfp[column_header],
        y=dfp['Nacional'],
        mode='markers',
        name=str(column_header)
    ))

# Update the layout

fig.update_layout(
    xaxis_title="Periodo",
    yaxis_title="Poblacion",
    title="Evolucion Empleo, Nacional segun las fechas",
)

fig.show()
```

Preprocesamiento Dataset

```python
dfp.rename(columns={'Área':'Urbana','Unnamed: 5':'Rural','Sexo':'Hombre','Unnamed: 7':'Mujer'}, inplace=True)
dfp=dfp.drop([0],axis=0)
```

```python
meses={"ene":"jan","feb":"feb","mar":"mar","abr":"apr","may":"may","jun":"jun","jul":"jul","ago":"aug","sep":"sep","oct":"oct","nov":"nov","dic":"dec"}
for i in range(len(dfp)):
  month_name=str(dfp.Periodo.iloc[i]).split('-')[0]
  month_name= meses.get(month_name)
  datetime_object = datetime.datetime.strptime(month_name, "%b")
  month_number = datetime_object.month
  year=str(dfp.Periodo.iloc[i]).split('-')[1]
  date_formated=("20"+str(year)+"-"+str(month_number)+"-01")
  dfp.Periodo.iloc[i]=date_formated
```

Agrupacion por Subempleo

```python
df=dfp.loc[dfp['Indicadores'].isin(['Subempleo'])]
df=df[['Periodo','Nacional',	'Urbana',	'Rural',	'Hombre',	'Mujer']]
dfp1=df['Periodo']
df=df.replace('\.','',regex=True)
dfp2=df.iloc[:,1:].apply(pd.to_numeric)
df=pd.concat([dfp1, dfp2], axis=1)
df.tail(5)
```

```python
fig = px.histogram(df, x="Periodo", y="Nacional", histfunc="avg", title="Histograma Subempleo Nacional")
fig.update_traces(xbins_size="M1")
fig.update_xaxes(showgrid=True, ticklabelmode="period", dtick="M4", tickformat="%b\n%Y")
fig.update_layout(bargap=0.1)
fig.add_trace(go.Scatter(mode="markers", x=df["Periodo"], y=df["Nacional"], name="Total de Personas"))

fig.show()
```

Conclusion

El subempleo en octubre del 2020es el mayor comparado con los demas muestras en la poblacion.

```python

df=dfp.loc[dfp['Indicadores'].isin(['Empleo Adecuado/Pleno'])]
df=df[['Periodo','Nacional',	'Urbana',	'Rural',	'Hombre',	'Mujer']]
dfp1=df['Periodo']
df=df.replace('\.','',regex=True)
dfp2=df.iloc[:,1:].apply(pd.to_numeric)
df=pd.concat([dfp1, dfp2], axis=1)
df.tail(5)

fig = px.histogram(df, x="Periodo", y="Nacional", histfunc="avg", title="Histograma Empleo Pleno Nacional")
fig.update_traces(xbins_size="M1")
fig.update_xaxes(showgrid=True, ticklabelmode="period", dtick="M4", tickformat="%b\n%Y")
fig.update_layout(bargap=0.1)
fig.add_trace(go.Scatter(mode="markers", x=df["Periodo"], y=df["Nacional"], name="Total de Personas"))

   
fig.show()
```
