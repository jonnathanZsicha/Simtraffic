```python
import random
import matplotlib.pyplot as plot
def sumaValores(a,b):
  c = a+b
  frecuencia(c)
  return c

vec = []
def frecuencia(c):
  vec.append(c)

def graficar(repeticiones):
  for i in range(0,repeticiones):
    dadoA = random.randint(1,6)
    dadoB = random.randint(1,6)
    sumaValores(dadoA,dadoB)
  mapa_vector = {}

  for suma in vec:
	  if suma in mapa_vector:
		  mapa_vector[suma] += 1
	  else:
		  mapa_vector[suma] = 1
  intervalos = range(min(vec), 12 + 2) 
  plot.hist(x=vec, bins=intervalos, color='#9AFF33', rwidth=0.85)
  w1 = 'sumas de dados '
  w2 = str(repeticiones)
  plot.title(w1 + w2)
  plot.xlabel('Sumas')
  plot.ylabel('Frecuencia')
  plot.xticks(intervalos)
  plot.show() 

graficar(10)
```
